```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
train = h2o.import_file('trainms.csv')
```

```python
test = h2o.import_file('testms.csv')
```

```python
x =train.columns
y = "treatment"
x.remove(y)
```

```python
train.columns
```

```python
# For binary classification, response should be a factor
train[y] = train[y].asfactor()
# test[y] = test[y].asfactor()
```

```python
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
aml.leader
```

```python
test.shape
```

```python
preds = aml.predict(test)
```

```python
final = pd.DataFrame(columns=["s.no","treatment"])
# final["id"] = train["s.no"]
# final["treatment"] = preds["predict"]
```

```python
test_df = test.as_data_frame()
pred_df = preds.as_data_frame()
```

```python
final["s.no"] = test_df["s.no"]
```

```python
final["treatment"] = pred_df["predict"]
```

```python
final.to_csv("solution.csv",index=False)
```
