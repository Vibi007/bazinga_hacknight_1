```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
train = h2o.import_file('clean_data_final.csv')
```

```python
imp_col = ["work_interfere","family_history","care_options","Gender","benefits","coworkers","obs_consequence","anonymity","leave"]
x =train.columns
x1=train[imp_col].columns
y = "treatment"
x.remove(y)
```

```python
train[y] = train[y].asfactor()
```

```python
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
aml_imp = H2OAutoML(max_models=20, seed=1)
aml_imp.train(x=x1, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

## Prediction

```python

test = h2o.import_file('clean_final_T.csv',header=1)
```

```python
test.columns
```

```python
preds = aml.predict(test)
```

```python
final = pd.DataFrame(columns=["s.no","treatment"])
test_df = test.as_data_frame()
pred_df = preds.as_data_frame()
final["s.no"] = test_df["s.no"]
final["treatment"] = pred_df["predict"]
final.to_csv("solution_h2O_final.csv",index=False)
```
